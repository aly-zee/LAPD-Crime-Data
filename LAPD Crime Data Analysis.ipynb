
Demographic Analysis on the lapd reprted crimes. Are there any demographic factors such as age, gender, income level that correlate with higher crime rates?

Dataset: https://data.lacity.org/
API/Web Scraping: Data for income levels through out LA County


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

crimes = pd.read_csv("lapd_crime_data.csv")
stations = pd.read_csv("LAPD_Police_Stations.csv")
income = pd.read_pickle('income_df.pkl')
```

```python
# converting date reported and date occ to date time format
crimes['DATE OCC'] = pd.to_datetime(crimes['DATE OCC'])
# crimes['Date Rptd'] = pd.to_datetime(crimes['Date Rptd'])
```

```python
crimes.dtypes
```

```python
# filtering the data to only include date of occ for 2022 since our income data is only for the year 2022
start_date = pd.to_datetime('2022-01-01')
end_date = pd.to_datetime('2022-12-31')

crimes = crimes[(crimes['DATE OCC']>=start_date)& (crimes['DATE OCC'] <= end_date)]

# resetting the index after filtering the data
crimes = crimes.reset_index(drop = True)

crimes
```

```python
# drop unnecessary columns
crimes.drop(['Crm Cd 2','Crm Cd 3', 'Crm Cd 4'],axis=1,inplace=True)
```

```python
crimes
```

```python
# get unique area names
crimes['AREA NAME'].unique()
crimes['AREA NAME'].nunique()
```

```python
stations

# make the division column lower case
stations['DIVISION'] = stations['DIVISION'].str.lower()
stations
```

```python
crimes['AREA NAME'] = crimes['AREA NAME'].str.strip().str.lower()
print(crimes['AREA NAME'].unique())
print(stations['DIVISION'].unique())
```

```python
# change n holywood to north hollywood, west la to west los angeles in the crimes df
crimes['AREA NAME'].replace({'n hollywood':'north hollywood','west la':'west los angeles'},inplace = True)
print(crimes['AREA NAME'].unique())
```

```python
stations.rename(columns={'DIVISION':'AREA NAME'},inplace=True)
# convert zip from int to object in stations DF
stations['Zip'] = stations['Zip'].astype(object)
```

```python
# rename the column 'zip_code' to 'Zip'
income.rename(columns = {'zip_code':'Zip'},inplace = True)
income['median_income'] = income['median_income'].astype(object)
income['AREA NAME'].nunique()
```

```python

# merging the crimes and stations DFs
crimesv1 = pd.merge(crimes,stations,how='left',on='AREA NAME')
crimesv1
# crimesv1['Zip'] = crimesv1['Zip'].astype(object)
new_crimes = pd.merge(crimesv1, income, how = 'left',on='AREA NAME')
# # new_crimes['median_income'].isna().sum()
new_crimes
# # crimesv1.dtypes
```

```python
new_crimes.drop('Zip_y',axis=1, inplace=True)
new_crimes
```

```python
area_name_counts = new_crimes['AREA NAME'].value_counts()
```

```python
area_name_counts_dict = area_name_counts.to_dict()
area_name_counts_dict
```

```python
# income_df['AREA NAME'] = income_df['zip_code'].map(zip_area)
income['crimes'] = income['AREA NAME'].map(area_name_counts)
```

```python
income['size'] = income['crimes']
```

```python
crime_data = income.copy()
crime_data.drop('size',axis=1,inplace=True)
crime_data
```

```python
sns.relplot(x='AREA NAME', y='median_income',data = income, size='crimes',hue='crimes')
plt.xticks(rotation = 90)
plt.xlabel("Los Angeles County Divisions")
plt.ylabel("Median Income")
plt.title("Crime per Division and Income ")
plt.show()
```
