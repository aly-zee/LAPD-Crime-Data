```python
import pandas as pd
import requests
url = "https://api.census.gov/data/2022/acs/acs5/subject?get=NAME,S1903_C03_001E&for=zip%20code%20tabulation%20area:*"
response = requests.get(url)
```

```python
if response.status_code == 200:
    income_data = response.json()
income_data
```

```python
type(income_data)
```

```python
income_data[0]
```

```python
# making the first row the columns
income_df = pd.DataFrame(income_data, columns = income_data[0])
# dropping the first row and resetting the row index
income_df = income_df.drop(0).reset_index(drop=True)
# drop the name column. Keep median income and zip code
income_df.drop('NAME',axis=1,inplace=True)
income_df
```

```python
# raname the columns
income_df.rename(columns = {'S1903_C03_001E':'median_income','zip code tabulation area':'zip_code'},inplace=True)
income_df.dtypes
```

```python
# converting the median_income to int
income_df['median_income'] = income_df['median_income'].astype(int)
income_df.dtypes
```

```python
zip_code =['90731','90061','90003','90066','90062','90011','90014','90025','90033','90019','90006','90017','90028','91401'
,'91601','91335','91304','91331','91325','91345','90065']

zip_area = {'90731':'harbor','90061':'southeast','90003':'77th street','90066':'pacific','90062':'southwest','90011':'newton','90014':'central','90025':'west los angeles','90033':'hollenbeck','90019':'wilshire','90006':'olympic','90017':'rampart','90028':'hollywood','90065':'northeast','91601':'north hollywood','91401':'van nuys','91335':'west valley','91304':'topanga','91331':'foothill','91325':'devonshire','91345':'mission','90065':'northeast'
}
income_df['AREA NAME'] = income_df['zip_code'].map(zip_area)
income_df = income_df[income_df['zip_code'].isin(zip_code)].reset_index(drop = True)
income_df
```

```python
income_df.to_pickle('income_df.pkl')
```
